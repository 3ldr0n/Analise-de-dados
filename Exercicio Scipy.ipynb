```python
import numpy as np

from matplotlib import pyplot as plt
```

## Geom√©trica, com probabilidade p = 1/3

```python
from scipy.stats import geom

fig, ax = plt.subplots()

p = 1/3
mean, var, skew, kurt = geom.stats(p, moments='mvsk')
x = np.arange(geom.ppf(0.01, p),
              geom.ppf(0.99, p))
ax.plot(x, geom.pmf(x, p), 'bo', ms=8, label='geom pmf')
ax.vlines(x, 0, geom.pmf(x, p), colors='b', lw=5, alpha=0.5)
plt.legend()
```

## Poisson, com mu = 0.4

```python
from scipy.stats import poisson

fig, ax = plt.subplots()

mu = 0.4
mean, var, skew, kurt = poisson.stats(mu, moments='mvsk')
x = np.arange(poisson.ppf(0.01, mu),
              poisson.ppf(0.99, mu))
ax.plot(x, geom.pmf(x, mu), 'bo', ms=8, label='poisson pmf')
ax.vlines(x, 0, poisson.pmf(x, mu), colors='b', lw=5, alpha=0.5)
plt.legend()
```

## Exponencial, com lambda = 0.5

```python
from scipy.stats import expon

fig, ax = plt.subplots()

_lambda = 0.5
mean, var, skew, kurt = expon.stats(_lambda, moments='mvsk')
x = np.linspace(expon.ppf(0.01),
                expon.ppf(0.99), 100)
ax.plot(x, expon.pdf(x),
       'r-', lw=5, alpha=0.6, label='expon pdf')
plt.legend()
```

## Normal padr√£o

```python
from scipy.stats import norm

fig, ax = plt.subplots()

mean, var, skew, kurt = norm.stats(moments='mvsk')
x = np.linspace(norm.ppf(0.01),
                norm.ppf(0.99), 100)
ax.plot(x, norm.pdf(x),
       'r-', lw=5, alpha=0.6, label='norm pdf')
```

## Uniforme no intervalo [0, 10]

```python
from scipy.stats import randint

fig, ax = plt.subplots()

low = 0
high = 10
mean, var, skew, kurt = randint.stats(low, high, moments='mvsk')
x = np.arange(randint.ppf(0.01, low, high),
              randint.ppf(0.99, low, high))
ax.plot(x, randint.pmf(x, low, high), 'bo', ms=8, label='randint pmf')
ax.vlines(x, 0, randint.pmf(x, low, high), colors='b', lw=5, alpha=0.5)
```
